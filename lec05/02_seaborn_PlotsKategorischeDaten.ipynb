# Diagramme fuer kategorische Daten

Hier schauen wir uns nun die Verwendung von Seaborn zur Darstellung kategorischer Daten an! Es gibt dafür einige wesentliche Plottypen:


* factorplot
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

Im Folgenden gibt es Beispiele fuer jeden einzelnen Plottypen!

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.describe()
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

## barplot and countplot

**Barplot** ist eine allgemeine Darstellung, die es ermöglicht, die kategorischen Daten basierend auf einer Funktion zu aggregieren, standardmäßig den Mittelwert:

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='class',y='survived',data=titanic)
```

```python
import numpy as np
```

Man kann den estimator Parameter zu einer anderen Funktion setzen:

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

### countplot

Dies ist im Wesentlichen dasselbe wie bei Barplot, außer dass der Schätzer explizit die Anzahl der Vorkommnisse zählt. Deshalb übergeben wir nur den x-Wert:

```python
sns.countplot(x='sex',data=tips)
```

## boxplot und violinplot

Boxplots und Violinplots werden verwendet, um die Verteilung der kategorischen Daten darzustellen. Ein Box-Plot (oder Box-and Whisker-Plot) zeigt die Verteilung quantitativer Daten so, dass Vergleiche zwischen Variablen oder über Ebenen einer kategorischen Variablen hinweg möglich sind. Die Box zeigt die Quartile des Datensatzes, während sich die Whisker erstrecken, um den Rest der Verteilung darzustellen, mit Ausnahme von Punkten, die nach einem Verfahren (Funktion des Interquartilbereichs), als "Ausreißer" bestimmt werden.

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
# Ganzes Dataframe mit orient='h'
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")
```

### violinplot

Eine Violinplot spielt eine ähnliche Rolle wie eine Box- und Whiskerplot. Es zeigt die Verteilung quantitativer Daten über mehrere Ebenen einer (oder mehrerer) kategorialer Variablen, so dass diese Verteilungen verglichen werden können. Im Gegensatz zu einem Boxplot, bei dem alle Plotkomponenten den tatsächlichen Datenpunkten entsprechen, verfügt der Violinplot über eine Kerneldichte-Schätzung der zugrunde liegenden Verteilung.

```python
sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=True,palette='Set1')
```

## stripplot und swarmplot

Der Stripplot zeichnet ein Scatterplot, bei dem eine Variable kategorisch ist. Ein Streifenplot kann für sich allein gezeichnet werden, ist aber auch eine gute Ergänzung zu einem Box- oder Geigenplot, wenn alle Beobachtungen zusammen mit einer Darstellung der zugrunde liegenden Verteilung dargestellt werden möchten.

Das Schwarmplot ist ähnlich wie stripplot(), aber die Punkte werden so angepasst (nur entlang der kategorischen Achse), dass sie sich nicht überlappen. Dies gibt eine bessere Darstellung der Verteilung der Werte, obwohl es nicht so gut zu einer großen Anzahl von Beobachtungen skaliert (sowohl in Bezug auf die Fähigkeit, alle Punkte darzustellen, als auch in Bezug auf die Berechnung, die für ihre Anordnung erforderlich ist).

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=0.4)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1',split=True)
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips, palette="Set1", split=True)
```

### Kategorische Plots kombinieren

```python
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
```

## catplot (ehemals: Factorplot)

Andere Möglichkeit: Catplot ist die allgemeine Form eines kategorischen Plots. Es kann ein **kind** Parameter übergeben werden, um den Plottyp anzupassen:

```python
sns.catplot(x='sex',y='total_bill',data=tips,kind='bar')
```
