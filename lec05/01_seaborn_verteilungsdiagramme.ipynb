# Verteilungsdiagramme (distribution plots)

Hier werden wir uns anschauen, wie man die Verteilung eines Datensatzes, veranschaulichen kann.

* distplot
* jointplot
* pairplot
* rugplot
* kdeplot

## Import

```python
import seaborn as sns
```

## Daten
Seaborn stellt Beispieldatensätze zur Verfügung, welche wir hier verwenden werden.

```python
tips = sns.load_dataset('tips') 
```

```python
tips.head()
```

## distplot

Der distplot zeigt die Verteilung einer univariaten (eindimensionalen) Menge von Beobachtungen.

```python
sns.distplot(tips['total_bill'])
# Warnung das Funktion bald entfernt wird
```

```python
sns.histplot(tips['total_bill'],kde=True,stat='density', linewidth=0)
```

Um die kde-Ebene (kernel density estiamtion) zu entfernen und einfach das Histogramm zu haben, setzt man den Parameter kde=False

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

## jointplot

jointplot() ermoeglicht es, zwei distplots für bivariate Daten zu vergleichen. Hierbei setzt man den **kind** Parameter, um anzugeben, welche Art von Plot man verwenden moechte: 

* “scatter” 
* “reg” 
* “resid” 
* “kde” 
* “hex”

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

## pairplot

pairplot zeichnet paarweise Beziehungen über ein gesamtes Dataframe (nur für die numerischen Spalten) und unterstützt ein Argument für den Farbton (für kategorische Spalten). 

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

## rugplot

Rugplots sind eigentlich ein sehr einfaches Konzept. Sie zeichnen lediglich eine Strichmarkierung für jeden Punkt einer univariaten Verteilung. Sie sind der Baustein eines KDE-Plots:

```python
sns.rugplot(tips['total_bill'])
```

## kdeplot

kdeplots sind [Kernel Density Estimation plots (Kerndichteschaetzer)*](https://de.wikipedia.org/wiki/Kerndichteschätzer). Diese KDE-Diagramme ersetzen jede einzelne Beobachtung durch eine Gaußsche (Normal-)Verteilung, die sich um diesen Wert dreht. Zum Beispiel:

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['tip'])
sns.rugplot(tips['tip'])
```
